```python
#import bokeh packages
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prep data
x = [1,2,3,4,5]
y = [1,2,3,4,5]

# figure object
fig = figure()

#output file for the plot
output_file("line.html")

#plot line graph
fig.line(x,y)

#write the graph 
show(fig)
```

```python
#import bokeh packages
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prep data
df = pandas.read_excel("weather.xlsx")
x = df["Temperature"]/10
y = df["Pressure"]/10
       
# figure object
fig = figure()

#output file for the plot
output_file("weather.html")

#plot line graph
fig.line(x,y)

#write the graph 
show(fig)
```
